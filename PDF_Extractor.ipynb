```python
resume_html = df['Resume_html']
length = len(df)
list7 = list(list())
list8 = list(list())
list9 = list()
for content in range(0,len(df)):
    soup = BeautifulSoup(df['Resume_html'][content], 'html.parser')
    # table = soup.find_all('div',{'class':'paragraph PARAGRAPH_NAME firstparagraph'})
    # list1 = list()
    # for row in table:
    #     list1.append(row.text)
    # list9.append(list1)

    temp = soup.findAll('div',{'class':'field singlecolumn'})
    length = len(temp)
    #print(length)
    list1 = Key_Extractor['Skills']
    if length >= 1:
        list2 = (temp[length-1].text).split(',')
    else:
        list2.append('NA')
    list1.append(list2)
    Key_Extractor['Skills'] = list1


    table = soup.findAll('span',{'class':['degree','companyname companyname_educ']})
    count = 0
    temp = 0
    list5 = []
    list6 = []
    for row in table:
        if count % 2 == 0:
            if row.text != ' ':
                list5.append(row.text)
            else:
                list5.append('NA')
        else:
            if row.text != ' ':
                list6.append(row.text)
            else:
                list6.append('NA')
        count += 1
    list7.append(list5)
    list8.append(list6)


    Key_Extractor['Category'] = df['Category'].tolist()
    Key_Extractor['Degree'] = list7
    Key_Extractor['Institution'] = list8


```
