```python
import pandas as pd
hugging_df = pd.read_csv('Selected_Company_Job _Description.csv',index_col='Unnamed: 0')
hugging_df
```

```python
from spacy.lang.en import English

# Load English tokenizer, tagger, parser, NER and word vectors
nlp = English()
list1 = []
for i in range(0,len(hugging_df)):
    text = hugging_df.loc[i,'job_description']

    #  "nlp" Object is used to create documents with linguistic annotations.
    my_doc = nlp(text)

    # Create list of word tokens
    token_list = []
    for token in my_doc:
        token_list.append(token.text)

    from spacy.lang.en.stop_words import STOP_WORDS

    # Create list of word tokens after removing stopwords
    filtered_sentence =[] 

    for word in token_list:
        lexeme = nlp.vocab[word]
        if lexeme.is_stop == False:
            filtered_sentence.append(word) 
    #print(token_list)
    #print(filtered_sentence)
    filtered_sentence = [i for a,i in enumerate(filtered_sentence) if i!=' ']
    list1.append(filtered_sentence)
```

```python
import ast
def string_to_list(string):
    res = ast.literal_eval(string)
    return res
```

```python
hugging_df['position_title'] = hugging_df['position_title'].apply(lambda x: [x])
```

```python
hugging_df['job_description_tokenization'] = [i for i in list1]
```

```python
def detail_skills(list1):
    sentence = ""
    count = 0
    length = len(list1)
    for i in range(0,length):
        sentence += list1[i]
        count += 1
        if count >=1 and count < length:
            sentence += " "
    return sentence
    
```

```python
actual_df['Detail_Skills']=actual_df['Skills'].apply(lambda x: detail_skills(x)) 
actual_df
```

```python
def acadmeic_detail():
    length = len(actual_df)
    for i in range(0,length):
        temp1 = actual_df.loc[i,'Degree']
        temp2 = actual_df.loc[i,'Institution']
        detail = ""
        for j in range(0,len(temp1)):
            temp = " "
            if temp1[j] == 'NA' or temp1[j] == 'NA':
                continue
            else:
                detail = temp1[j] + ' from ' + temp2[j] + ','+' '
        actual_df.loc[i,'Acadmeic'] = detail
```

```python
import torch
from transformers import DistilBertConfig, DistilBertModel, DistilBertTokenizer
```

```python
new_max_position_embeddings = 1024  # Change this to your desired value

# Create a DistilBERT configuration with the new max_position_embeddings
config = DistilBertConfig.from_pretrained('distilbert-base-uncased', max_position_embeddings=new_max_position_embeddings)

# Load the DistilBERT model with the modified configuration
model = DistilBertModel(config=config)

# Create a DistilBERT tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')


for i in range(0,len(actual_df)):

    # Tokenize and encode the text
    text = actual_df.loc[i,'Candidate_Summary']
    # Tokenize and encode the text
    # print(text)

    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=new_max_position_embeddings)

# Forward pass through the model to get embeddings
    with torch.no_grad():
        outputs = model(**inputs)

# Extract the embeddings from the model outputs
    embeddings = outputs.last_hidden_state
    actual_df.loc[i,'Embedding_Candidate_Summary'] = embeddings
```

```python
import numpy as np
def dimension_change(vector1):

    # Define the original vector with dimensions 1x71x768
    original_vector = vector1  # Replace with your actual data or initialization method

    # Define the desired dimensions
    desired_dimensions = (1, 1024, 768)

    # Create a new tensor with the desired dimensions filled with zeros
    padded_vector = np.zeros(desired_dimensions)

    # Copy the original data into the padded tensor
    padded_vector[:, :original_vector.shape[1], :] = original_vector

    return padded_vector
```
