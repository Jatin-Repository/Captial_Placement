```python
import pandas as pd
import numpy as np
```

```python
job_desc = pd.read_csv('training_data.csv')
job_desc
```

```python
import re

def remove_special_characters(text):
    pattern = r'[!@#$%^&*()_+{}\[\]:;<>,.?~\\|/|\n|\t]'
    cleaned_text = re.sub(pattern, ' ', text)
    return cleaned_text
```

```python
job_desc['job_description'] = job_desc['job_description'].apply(lambda x: remove_special_characters(x))
```

```python
job_desc
```

```python
job_data = pd.DataFrame(job_desc.sample(20)) # random job description
```

```python
job_data=job_data.reset_index()
```

```python
job_data.drop(columns='index',inplace=True)
```

```python
job_data.to_csv('Selected_Company_Job _Description.csv')
```
